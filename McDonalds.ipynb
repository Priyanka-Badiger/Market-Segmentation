**Ten Steps of Market Segmentation Analysis**

---



**Step 1: Deciding (not) to Segment**

The organization's willingness and ability to make significant changes goes hand in hand with its commitment to market segmentation. One of the truisms of segmentation strategy is that using the scheme must be more profitable than marketing without it, after deducting the cost of developing and implementing the scheme. The strategy for market segmentation must be developed at the highest executive level.

**Implementation Barriers**: The first group of barriers relates to senior management. A second group of barriers relates to organisational culture. The lack of a formal marketing function or, at the very least, a qualified marketing expert in the organisation is closely related to these barriers. Another impediment may be the organization's objective constraints, such as a lack of financial resources or the inability to make the necessary structural changes. If barriers cannot be removed, the option of abandoning market segmentation as a potential future strategy should be seriously considered.

---



**Step 2: Specifying the Ideal Target
Segment**

**Knock-Out Criteria**: Determine whether market segments identified through market segmentation analysis are eligible to be evaluated using segment attractiveness criteria. The segment must be homogeneous, distinct, large enough, matching, identifiable and reachable. 

**Attractiveness Criteria**: The criteria for attractiveness are not binary in nature. Segments are not graded as either meeting or failing to meet attractiveness criteria. Rather, each market segment is rated separately.

**Implementing a Structured Proces**: The segmentation team determines the segment attractiveness and organisational competitiveness values. It is recommended that no more than six factors be used to calculate these criteria. If a core team of two to three people is primarily in charge of market segmentation analysis, this team could propose an initial solution and report their choices to an advisory committee made up of representatives from all organisational units for discussion and possible modification. Knowing precisely what information about market segments is important to the organisation ensures that all information is captured when data is collected.
The typical approach to weighting is to assign 100 points to each segmentation criterion to all team members. These allocations must then be negotiated until a consensus is reached.

---



**Step 3: Collecting Data**

**Segmentation Variables**: The variable in empirical data used in commonsense segmentation to split the sample into market segments is referred to as the segmentation variable. Descriptor variables are used to describe the segments in detail. The correct description allows for the development of a customised product, the determination of the most appropriate pricing strategy, the selection of the best distribution channel, and the most effective communication channel for advertising and promotion. Empirical data for segmentation studies can come from a variety of sources, including survey studies.

**Segmentation Criteria**: The term segmentation criterion refers to the type of data used for market segmentation. It can also refer to a single concept, such as benefits sought. Requires prior knowledge about the market. Profitability, bargaining power, preferences for benefits or products, barriers to choice, and consumer interaction effects are the most important differences between consumers in terms of market segmentation. 


1.   **Geographic Segmentation**: The consumer's place of residence is the only criterion used to define market segments. Living in the same country or region does not always imply that people share other characteristics that are important to marketers.
2.   **Socio-Demographic Segmentation**: Age, gender, income, and education are common socio-demographic segmentation criteria. In many cases, the socio-demographic criterion is not the root cause of product preferences, resulting in insufficient market insight for optimal segmentation decisions.
3.  **Psychographic Segmentation**: Refers to the grouping of people based on psychological criteria such as their beliefs, interests, preferences, aspirations, or benefits sought when purchasing a product. It is difficult to identify a single personality trait that will provide insight into the psychographic dimension of interest. 
4.  **Behavioural Segmentation**: Prior experience with the product, frequency of purchase, amount spent on purchasing the product on each occasion (or across multiple purchase occasions), and information search behaviour are all examples of possible behaviours. It is limits oneself to the study of the organization's existing customers.

**Data from Survey Studies**: Survey data is inexpensive and simple to collect, making it a viable approach for any organisation, but it can be contaminated by a variety of biases. 


1.   **Choice of Variables**: All variables relevant to the construct captured by the segmentation criterion must be included in data-driven segmentation. Simultaneously, unnecessary variables must be avoided.The advice is to ask all necessary and unique questions while resisting the urge to include unnecessary or redundant questions. 
2.   **Response Options**: Nominal variables correspond to options that allow respondents to choose an answer from a list of unordered categories. Metric data allow for the execution of any statistical procedure, making them ideal for segmentation analysis. The options are ordered, so the data is ordinal. The visual analogue scale allows respondents to indicate a position along a continuous line between two endpoints, yielding metric data.
3.  **Response Styles**: Biases in survey data are common. A response bias is a systematic tendency to respond to a variety of questionnaire items based on criteria other than the specific item content. Survey responses reveal a wide range of response styles, including respondents' proclivity to use extreme answer options, the midpoint, and to agree with all statements.
4.  **Sample Size**:  data used in market segmentation analyses should
*   contain no unnecessary items
*   contain no correlated items
*   contain high-quality responses
*   be binary or metric
*   be free of response styles
*   include responses from a suitable sample given the aim of the segmentation
study
*   include a sufficient sample size given the number of segmentation variables (100 times the number of segmentation variables)

**Data from Internal Sources**: represent actual consumer behaviour. It may be systematically skewed due to overrepresentation of existing customers.

**Data from Experimental Studies**: Experiment data can come from both field and laboratory experiments. Conjoint studies and choice experiments yield data on the extent to which each attribute and attribute level influences choice.

**Principal Components Analysis**: Principal components analysis (PCA) transforms a multivariate data set containing metric variables to a new data set with variables – referred to as principal components – which are uncorrelated and ordered by importance. The first variable (principle component) contains most of the variability, the second principle component contains the second most variability, and so on.
Principal components analysis works off the covariance or correlation matrix
of several numeric variables. If all variables are measured on the same scale, and have similar data ranges, it is not important which one to use. If the data ranges are different, the correlation matrix should be used (which is equivalent to standardising the data)

**Step 6: Profiling Segments**

**Identifying Key Characteristics of Market Segments**: The aim of the profiling step is to get to know the market segments resulting from the extraction step. Profiling is only required when data-driven market segmentation is used. For commonsense segmentation, the profiles of the segments are predefined. At the profiling stage, we inspect a number of alternative market segmentation solutions. This is particularly important if no natural segments exist in the data, and either a reproducible or a constructive market segmentation approach has to be taken. Good profiling is the basis for correct interpretation of the resulting segments. Correct interpretation, in turn, is critical to making good strategic marketing decisions.

**Traditional Approaches to Profiling Market Segments**: Data-driven segmentation solutions are usually presented to users (clients, managers) in one of two ways: (1) as high level summaries simplifying segment characteristics to a point where they are misleadingly trivial, or (2) as large tables that provide, for each segment, exact percentages for each segmentation variable.

**Segment Profiling with Visualisations**: Neither the highly simplified, nor the very complex tabular representation typically used to present market segmentation solutions make much use of graphics, although data visualisation using graphics is an integral part of statistical data analysis. In addition, in times of big and increasingly bigger data, visualisation offers a simple way of monitoring developments over time.

**Identifying Defining Characteristics of Market Segments**: A good way to understand the defining characteristics of each segment is to produce a segment profile plot. The segment profile plot shows – for all segmentation variables – how each market segment differs from the overall sample.

**Assessing Segment Separation**: Segment separation can be visualised in a segment separation plot. The segment separation plot depicts – for all relevant dimensions of the data space – the overlap of segments. Segment separation plots are very simple if the number of segmentation variables is low, but become complex as the number of segmentation variables increases. But even in such complex situations, segment separation plots offer data analysts and users a quick overview of the data situation, and the segmentation solution.


```python
features = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']
X = data[features]

pca = PCA(n_components=2)
components = pca.fit_transform(X)

loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

fig = px.scatter(components, x=0, y=1, color=data['Like'])

for i, feature in enumerate(features):
    fig.add_shape(
        type='line',
        x0=0, y0=0,
        x1=loadings[i, 0],
        y1=loadings[i, 1]
    )
    fig.add_annotation(
        x=loadings[i, 0],
        y=loadings[i, 1],
        ax=0, ay=0,
        xanchor="center",
        yanchor="bottom",
        text=feature,
    )
fig.show()
```
